```python
%matplotlib inline

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
def logToMsgs(name):
    file = open(name)
    comments = 0
    while True:
        line = file.readline()
        if not line:
            break
        elif not line.startswith('#'):
            arr = line.strip().split(',')
            yield [(int(x) % 256) for x in arr[:-1]]
            
```

```python
def plotStats(num, name):
    msgs = list(logToMsgs('../resources/logs/'+name+'.log'))
    
    sizes = np.array([len(msg) for msg in msgs])
    codes = np.array([msg[0] for msg in msgs])
    ids = np.array([msg[0] for msg in msgs])#Customer IDs, which we know are the first params in all thingml messages in non-diversified versions
    
    plt.figure(num,figsize=(15,7))
    plt.subplot(1,3,1)
    plt.hist(sizes)
    plt.title("Sizes")

    plt.subplot(1,3,2)
    plt.hist(codes,bins=np.arange(0,257)-.5)
    plt.title("Codes")
    
    plt.subplot(1,3,3)
    plt.hist(ids,bins=np.arange(0,257)-.5)
    plt.title("IDs")
```
