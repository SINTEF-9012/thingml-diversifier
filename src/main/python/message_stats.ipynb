```python
%matplotlib inline

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
def logToMsgs(name):
    file = open(name)
    comments = 0
    while True:
        line = file.readline()
        if not line:
            break
        elif not line.startswith('#'):
            arr = line.strip().split(',')
            #yield [int(x) if int(x) >= 0 else 256+int(x) for x in arr[:-1]]
            yield [int(x) for x in arr[:-1]]
            
```

```python
def plotStats(num, name):
    msgs = list(logToMsgs('../resources/logs/'+name+'.log'))
    
    sizes = np.array([len(msg) for msg in msgs])
    id = np.array([msg[0] for msg in msgs])
    
    plt.figure(num,figsize=(15,7))
    plt.subplot(1,2,1)
    plt.hist(sizes)
    plt.title("Sizes")

    plt.subplot(1,2,2)
    plt.hist(id)
    plt.title("IDs")
```

```python
plotStats(1,'4.1')
plotStats(2,'4.2')
plotStats(3,'4.3')
plotStats(4,'4.4')
```
