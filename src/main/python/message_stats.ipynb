```python
%matplotlib inline

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
def logToMsgs(name):
    file = open(name)
    comments = 0
    while True:
        line = file.readline()
        if not line:
            break
        elif not line.startswith('#'):
            arr = line.strip().split(',')
            #yield [int(x) if int(x) >= 0 else 256+int(x) for x in arr[:-1]]
            yield [int(x) for x in arr[:-1]]
            
```

```python
def plotStats(num, name):
    msgs = list(logToMsgs('../resources/logs/'+name+'.log'))
    
    sizes = np.array([len(msg) for msg in msgs])
    id = np.array([msg[0] for msg in msgs])
    
    #In principles, we should do that for all params to figure out if some specific bytes tend to appear at some specific places 
    #We just know that msg[1] is the customer ID, and want to check if ID is shuffled when we diversify
    id2 = np.array([msg[1] for msg in msgs])
    
    
    plt.figure(num,figsize=(15,7))
    plt.subplot(2,2,1)
    plt.hist(sizes)
    plt.title("Sizes")

    plt.subplot(2,2,2)
    plt.hist(id)
    plt.title("Codes")
    
    plt.subplot(2,2,3)
    plt.hist(id2)
    plt.title("IDs")
```

```python
plotStats(1,'0.1')
plotStats(2,'0.2')
plotStats(3,'0.3')
plotStats(4,'0.4')

plotStats(5,'4.1')
plotStats(6,'4.2')
plotStats(7,'4.3')
plotStats(8,'4.4')

plotStats(9,'4.5-new')
```
